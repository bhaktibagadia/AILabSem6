Aim: Write a program to solve Tic-Tac-Toe using min-max algorithm

```python
def check_winner(board):
    # Checking rows
    for row in board:
        if row.count(row[0]) == len(row) and row[0] != " ":
            return True

    # Checking columns
    for col in range(len(board)):
        if all(board[row][col] == board[row+1][col] == board[row+2][col] != " " for row in range(len(board)-2)):
            return True

    # Checking diagonals
    if all(board[i][i] == board[i+1][i+1] == board[i+2][i+2] != " " for i in range(len(board)-2)):
        return True
    if all(board[i][len(board)-1-i] == board[i+1][len(board)-2-i] == board[i+2][len(board)-3-i] != " " for i in range(len(board)-2)):
        return True

    return False

def board_full(board):
  winner = check_winner(board)
  if not winner:
    count=0
    for i in range(0,len(board)):
      for j in range(0,len(board[i])):
        if board[i][j]=='X' or board[i][j]=='O':
          count+=1
    if count==9:
      return True
  return False
```

```python
def get_turn(board):
  x,o=0,0
  for i in range(0,len(board)):
    for j in range(0,len(board[i])):
      if board[i][j]=='X':
        x+=1
      elif board[i][j]=='O':
        o+=1
  if x>o:
    return False
  return True
```

```python
import copy
def print_board(board):
    print("*************************")
    for row in board:
        print(" | ".join(row))
        print("-" * 10)


li = []
li2 = []
maxDepth = float('-inf')

def dfs(board, maximizing_player,depth):
    global maxDepth
    maxDepth = max(maxDepth,depth)
    if check_winner(board):
        if maximizing_player:
            return -1
        else:
            return 1
    elif board_full(board):
        return 0  # Draw


    if maximizing_player:
        max_eval = float('-inf')
        for i in range(len(board)):
            for j in range(len(board[i])):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    depth+=1
                    l = [copy.deepcopy(board)]
                    eval = dfs(board, False,depth)
                    l.append(eval)
                    l.append(depth)
                    li.append(l)
                    depth-=1
                    board[i][j] = ' '
                    max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for i in range(len(board)):
            for j in range(len(board[i])):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    depth+=1
                    l = [copy.deepcopy(board)]
                    eval = dfs(board, True,depth)
                    l.append(eval)
                    l.append(depth)
                    li.append(l)
                    depth-=1
                    board[i][j] = ' '
                    min_eval = min(min_eval, eval)
        return min_eval


input_board = [['O', 'O', 'X'],
               [' ', 'X', ' '],
               ['O', 'X', ' ']]


current_player_is_x = get_turn(input_board)
best_score = dfs(input_board, current_player_is_x,0)
print("Best score:", best_score)
print('Initial State:')
print_board(input_board)
print('\n\n')

for i in range(1, maxDepth + 1):
    print("depth:",i)
    for l in li:
        if l[2] == i:
            print_board(l[0])
            print('Value:', l[1])
    print("\n\n")
```
